```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math
import itertools
```

```python
x0 = [0.5,0]
```

```python
mu = 1.5
```

```python
def van_der_pol(x,t):
    A = x[0]
    B = x[1]
    dx = np.zeros(2)
    dx[0] = B
    dx[1] = -mu*(A**2-1)*B-A
    return dx
```

```python
#solving odes
t0 = np.array(np.linspace(0,50,5000))

x = np.array(odeint(van_der_pol,x0,t0))

A0 = x[:,0]
B0 = x[:,1]
```

```python

plt.plot(A0,B0)
plt.xlabel("x")
plt.ylabel("dx/dt")
plt.show()
```

```python
plt.plot(t0,A0)
plt.xlabel("time")
plt.ylabel("x")
plt.show()
```
